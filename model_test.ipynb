```python
import tensorflow as tf
import chess
import chess.engine
import numpy as np

# load your saved keras model
model = tf.keras.models.load_model("mymodel_100epoch_9k.keras")
```

```python
piece_map = {
    "P": 1, "N": 2, "B": 3, "R": 4, "Q": 5, "K": 6,   # White pieces
    "p": -1, "n": -2, "b": -3, "r": -4, "q": -5, "k": -6,  # Black pieces
}

def board_to_input(board):
    arr = np.zeros((8, 8))
    for square, piece in board.piece_map().items():
        row = 7 - (square // 8)
        col = square % 8
        arr[row][col] = piece_map[piece.symbol()]
    return arr
```

```python
def predict_move(board):
    moves = list(board.legal_moves)
    best_move = None
    best_score = -9999

    for move in moves:
        board.push(move)
        x = board_to_input(board).reshape(1, 8, 8, 1)  # depends on your training
        score = model.predict(x, verbose=0)[0][0]      # adjust indexing for your output
        if score > best_score:
            best_score = score
            best_move = move
        board.pop()
    
    return best_move
```
